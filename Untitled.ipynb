```r
library(triangle)
library(igraph)
library(mvtnorm)

#numero de cenarios
n_cenarios <- 3000
gera_grafo <- function(){
  g <- make_graph(c(1,6,1,4,2,7,4,7,5,7,3,5))
  plot(g)
  letras <- c('A','B','C','D','E','F')
  p <- all_simple_paths(g,1, to=7)
}
gera_grafo()
matriz_correlacoes <- function(c1, c2, n,tipo){
  #matriz de correlacoes
  mc = rbind(c(1, 0, 0, 0, 0, 0, 0),
             c(0, 1, 0, c2, 0, 0, 0),
             c(0, 0, 1, 0, 0, 0, 0),
             c(0, c2, 0, 1, 0, 0, 0),
             c(0, 0, 0, 0, 1, c1, 0),
             c(0, 0, 0, 0, c1, 1, 0),
             c(0, 0, 0, 0, 0, 0, 1)
  )
  
  #medias para a normal multivariada
  medias <- rep(0,7)
  
  #1 - Gerar as Normais correlacionadas (Utilizando rmvnorm)
  x = rmvnorm(n,mean=medias,sigma=mc)
  
  #2 -  U <- Pnorm(c)
  u <- pnorm(x)
  if(tipo == 'custo'){
    #3 - Gera a matriz de custos
    a1 <- qtriangle(u[,1], 75, 110, 90)
    a2 <- qtriangle(u[,2], 120, 190, 150)
    a3 <- qtriangle(u[,3], 60, 85, 75)
    a4 <- qtriangle(u[,4], 45, 80, 60)
    a5 <- qtriangle(u[,5], 160, 240, 200)
    a6 <- qtriangle(u[,6], 210, 300, 250)
    a7 <- qtriangle(u[,7], 85, 140, 120)
    custos <- matrix(c(a1,a2,a3,a4,a5,a6,a7), ncol=7)
    custos
  }else if(tipo == 'prazo'){
    #gera a matriz de durações
    d1 <- qtriangle(u[,1], 1, 4, 2)
    d2 <- qtriangle(u[,2], 5, 7, 6)
    d3 <- qtriangle(u[,3], 2, 5, 4)
    d4 <- qtriangle(u[,4], 1, 4, 3)
    d5 <- qtriangle(u[,5], 4, 7, 5)
    d6 <- qtriangle(u[,6], 3, 5, 4)
    d7 <- qtriangle(u[,7], 1, 3, 2)
    prazos <- matrix(c(d1,d2,d3,d4,d5,d6,d7), ncol=7)
    prazos
  }
}

risco_custo <- function(n_cenarios=3000){
  
  #correlacao entre E e F
  corr1 <- 0.7
  #correlacao entre B e D
  corr2 <- 0.4
  
  #probabilidades da atividade f ocorrer para cada um dos cenarios
  prob_f <- rbinom(n_cenarios, 1, 0.25)
  
  #matriz com os custos de cada atividade para cada cenario
  custos_atividades <- matriz_correlacoes(corr1, corr2, n_cenarios,tipo='custo')
  
  #matriz com os custos da atividade F alterados para 0 se ela não ocorre ou inalterados
  #se ela ocorre
  custos_atividades[,6] = custos_atividades[,6] * prob_f
  
  #adiciona uma coluna a matriz anterior com o calculo do custo total por cenario
  custos_total <- cbind(custos_atividades, apply(custos_atividades, 1, sum))
  
  #pdf custo
  pdf <- density(custos_total[,8])
  plot(pdf, main='distribuicao de probabilidade do custo total')
  #cdf do custo total com o ponto onde o risco é de 15%
  cdf <- ecdf(custos_total[,8])
  
  plot(cdf, main='Distribuicao cumulativa do custo total')
  
  #grafico custo
  cor_custos_total <- cor(custos_total[,8], custos_total[,1:7],
                          method="spearman")
  cor_custos_total <- cor_custos_total/sum(cor_custos_total)
  
  o_ctotal <- order(cor_custos_total)
  
  ctotal_name <- seq(1, 7)[o_ctotal]
  
}

risco_prazo <- function(n_cenarios=3000){
    
    #correlacao entre E e F
    corr1 <- 0.7
    #correlacao entre B e D
    corr2 <- 0.4
    
    #probabilidades da atividade f ocorrer para cada um dos cenarios
    prob_f <- rbinom(n_cenarios, 1, 0.25)
    
    #matriz com os custos de cada atividade para cada cenario
    prazos_atividades <- matriz_correlacoes(corr1, corr2, n_cenarios,tipo='prazo')
    
    #matriz com os custos da atividade F alterados para 0 se ela não ocorre ou inalterados
    #se ela ocorre
    prazos_atividades[,6] = prazos_atividades[,6] * prob_f
    
    #adiciona uma coluna a matriz anterior com o calculo do custo total por cenario
    prazos_total <- cbind(prazos_atividades, apply(prazos_atividades, 1, sum))
    
    #pdf do custo total
    pdf <- density(prazos_total[,8])
    plot(pdf, main='distribuicao de probabilidade do prazo total')
    #cdf do custo total com o ponto onde o risco é de 15%
    cdf <- ecdf(prazos_total[,8])
    
    plot(cdf, main='Distribuicao cumulativa do prazo total')
    
    #gerando grafico de prazo
    prazo_total <- cor(prazos_total[,8], prazos_total[,1:7],
                            method="spearman")
    prazo_total <- prazo_total/sum(prazo_total)
    
    o_ptotal <- order(prazo_total)
    
    ptotal_name <- seq(1, 7)[o_ptotal]
    
  }
risco_custo()
risco_prazo()
```
